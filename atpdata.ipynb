```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn import linear_model
```

## Correlation of age vs aces/game

```python
df = pd.read_csv('./atp_matches_2024.csv')
df = df[['winner_age', 'w_ace']].dropna()
```

```python
df.info()
```

```python
X = np.array(df['winner_age'])
Y = np.array(df['w_ace'])
```

```python
fig, ax = plt.subplots(figsize=(12, 5))
ax.scatter(X, Y, color='mediumblue', s=20)
ax.set_xlabel('Age')
ax.set_ylabel('Aces/game')

plt.show()
```
